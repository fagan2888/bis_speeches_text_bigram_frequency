```python
from config import parameters
from utils import *
from collections import defaultdict
import statistics
```

```python
target_doc_list_filepath = 'target-list-units-wo-outlier_20191028-21-50-55.pkl'
quarterly_filepath = 'target-list-units-grouped_quarterly_20191028-21-52-08.pkl'
semiannually_filepath = 'target-list-units-grouped_semiannually_20191028-21-52-09.pkl'
annually_filepath = 'target-list-units-grouped_annually_20191028-21-52-09.pkl'
```

```python
target_doc_list = load_pkl(os.path.join(parameters.output_base_dir, target_doc_list_filepath))
len(target_doc_list)
```

### by author

```python
author_list = ['Greenspan', 'Bernanke', 'Yellen', 'Powell']
_doc_dict = defaultdict(lambda: 0)
_word_dict = defaultdict(lambda: 0)
for doc in target_doc_list:
    for author in author_list:
        if author in doc['short_info']:
            _doc_dict[author] += 1
            _word_dict[author] += len(doc['unigrams_by_sentence'])
            break

print('# of documents')
_sum = 0
for key, val in _doc_dict.items():
    print(key, '\t', val)
    _sum += val
print(_sum)

print('\n # of words')
_sum = 0
for key, val in _word_dict.items():
    print(key, '\t', val)
    _sum += val
print(_sum)
```

```python
sentence_length_list = list()

for _doc in target_doc_list:
    for i in range(len(_doc['sentences'])):
        unigrams_len = len(_doc['unigrams_by_sentence'][i])
        sentence_length_list.append(unigrams_len)

print('median:', statistics.median(sentence_length_list))
print('mean:', statistics.mean(sentence_length_list), '\n')

sentence_length_dict = defaultdict(lambda: list())
for _doc in target_doc_list:
    for i in range(len(_doc['sentences'])):
        unigrams_len = len(_doc['unigrams_by_sentence'][i])
        sentence_length_dict[unigrams_len].append({'sentence': _doc['sentences'][i], 'unigrams': _doc['unigrams_by_sentence'][i]})
print(sorted(sentence_length_dict.keys())[:4])
print(sorted(sentence_length_dict.keys())[-4:])
print("\n")

for i in sorted(sentence_length_dict.keys())[:4]:
    print('length=', i)
    for j in range(3):
        print(sentence_length_dict[i][j])
    print('\n')

for i in sorted(sentence_length_dict.keys())[-4:]:
    print('length=', i)
    print(sentence_length_dict[i])
    print('\n')

print('='*5, '\n sentence length')    
for _sent_len in sorted(sentence_length_dict.keys()):
    print(_sent_len, ',', len(sentence_length_dict[_sent_len]))
```

## by period

```python
quarterly_target_doc_dict = load_pkl(os.path.join(parameters.output_base_dir, quarterly_filepath))
print(len(quarterly_target_doc_dict.keys()))
semiannually_target_doc_dict = load_pkl(os.path.join(parameters.output_base_dir, semiannually_filepath))
print(len(semiannually_target_doc_dict.keys()))
annually_target_doc_dict = load_pkl(os.path.join(parameters.output_base_dir, annually_filepath))
print(len(annually_target_doc_dict.keys()))
```

```python
print('# of words by period')
for period_dict in [quarterly_target_doc_dict, semiannually_target_doc_dict, annually_target_doc_dict]:
    print('='*5, '\n')
    word_dict = defaultdict(lambda: 0)
    for _period in period_dict.keys():
        for doc in period_dict[_period]:
            word_dict[_period] += len(doc['unigrams_by_sentence'])
    for _period in sorted(word_dict.keys()):
        print(_period, ',', word_dict[_period])
```
